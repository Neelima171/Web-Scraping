```python
productname = [] 
price = []
rating = []
features = []
RR=[]
#pagenum = []

for i in range(1,21):
    start_time = time.time()
    url='https://www.flipkart.com/search?q=tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'+str(i)
    page = requests.get(url)
    page_content = page.text
    soup = BeautifulSoup(page_content)
    
    parent_div=soup.find_all('div', attrs={'class':'_3pLy-c row'})
    for x in parent_div:
        pname=x.find('div', attrs = {'class':'_4rR01T'})
        cost=x.find('div', attrs = {'class':'_1_WHN1'})
        rat=x.find('div', attrs = {'class':'_3LWZlK'})
        specs=x.find('ul', attrs = {'class':'_1xgFaf'})
        rr=x.find('span',attrs={'class':'_2_R_DZ'})
        if pname is None:
            productname.append(np.NaN)
        else:
            productname.append(pname.text)
            
        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if rat is None:
            rating.append(np.NaN)
        else:
            rating.append(rat.text)
            
        if specs is None:
            features.append(np.NaN)
        else:
            features.append(specs.text)
            
        if rr is None:
            RR.append(np.NaN)
        else:
            RR.append(rr.text)
                                 
      #  pagenum.append(i)
   # print('Page {} completed in {} seconds'.format(i, time.time()-start_time))
```

#### OBSERVATIONS: 
#### Here we gathering information about  different TV's from 20 different pages.

### Extracting TV name, brand and screen_size:

```python
tnb=[]
for i in range(1,21):
    tnb=productname
```

```python
len(tnb)
```

```python
tnb=list(map(lambda x: x.strip().replace('\n',' '),tnb))
print(len(tnb))
print(tnb)
```

```python
tbrand=[]
tname=[]
SS_cm=[]
SS_inch=[]
for i in tnb:
    B=re.findall(r'(\w+).*',i)
    N=re.findall(r'\w*(\w*.*)',i)
    ss1=re.findall(r'(\d+\d+) cm',i)
    ss1=["0" if x =='' else x for x in ss1]
    ss2=re.findall(r'(\d+\d+) inch',i)
    ss2=["0" if x =='' else x for x in ss2]
    if len(B)==0:
        tbrand.append(np.nan)
    else:
        tbrand.append(B[0])
    if len(N)==0:
        tname.append(np.nan)
    else:
        tname.append(N[0])
    if len(ss1)==0:
        SS_cm.append(np.nan)
    else:
        SS_cm.append(ss1[0])
    if len(ss2)==0:
        SS_inch.append(np.nan)
    else:
        SS_inch.append(ss2[0])
print(len(tname))
print(tname)
print(len(tbrand))
print(tbrand)
print(len(SS_cm))
print(SS_cm)
print(len(SS_inch))
print(SS_inch)
```

#### OBSERVATIONS:
#### Here we extracted information about TV_Brand, Name and Screensize in cms and inches.
#### And for null values we appended then with NaN.

### Extracting warranty and soundoutput:

```python
f=[]
for i in range(1,21):
    f=features
```

```python
Wat=[]
So=[]
for i in f:
    w=re.findall(r'(\d|\d.\d) Year',i)
    so=re.findall(r'(\d\d) W',i)
    if len(w)==0:
        Wat.append(np.nan)
    else:
        Wat.append(w[0])
    if len(so)==0:
        So.append(np.nan)
    else:
        So.append(so[0])
print(len(Wat))
print(Wat)
print(len(So))
print(So)
```

#### OBSERVATIONS:
#### Here we extracted information about Warranty and Soundoutput of TV's.
#### And we have not got any null values to appended them with NaN values.

### Extracting Rating and Reviews:

```python
R=[]
for i in range(1,21):
    R=R+parent_div
```

```python
rr = []
for i in R:
    r = i.find_all('span',class_='_2_R_DZ')
    if len(r) == 0:
        rr.append('')
    else:
        r = r[0].text
        rr.append(r)
        
print(len(rr))
print(rr)
```

```python
Rat=[]
Rew=[]
for i in rr:
    Ra=re.findall(r'(\d*|\d*|\d*[,]\d*[,]\d*|\d*[,]\d*) Ratings',i)
    Re=re.findall(r'(\d*|\d*|\d*[,]\d*[,]\d*|\d*[,]\d*) Reviews',i)
    if len(Ra)==0:
        Rat.append(np.nan)
    else:
        Rat.append(Ra[0])
    if len(Re)==0:
        Rew.append(np.nan)
    else:
        Rew.append(Re[0])
        
print(len(Rat))
print(Rat)
print(len(Rew))
print(Rew)
```

#### OBSERVATIONS:
#### We extracted how many people rated and reviewed the TV's.
#### And for null values we appended them with NaN values.

### Creating a dataframe:

```python
tv_df=pd.DataFrame({'TV_Brand':tbrand,'TV_Name':tname,'Star_Rating':rating,'Features':features,'Warranty':Wat,
                    'ScreenSize_cm':SS_cm,'ScreenSize_inch':SS_inch,'Sound_Output':So,'Rating':Rat,'Reviews':Rew,
                    'Price':price})
```

```python
tv_df.head(2)
```

### Extracting Supporting_apps, Resolution of TV's:

```python
my_regex=r'^[\D]+Youtube'
tv_df['Supported_apps']=tv_df['Features'].apply(lambda x:re.compile(my_regex).findall(x))
#tv_df['Supported_apps'].head()
```

```python
my_regex=r'[\d+][\s0-9A-Za-z]+Pixels'
tv_df['Resolution']=tv_df['Features'].apply(lambda x:re.compile(my_regex).findall(x))
#tv_df['Resolution'].head()
```

```python
tv_df['Supported_apps']=tv_df['Supported_apps'].apply(lambda x:''.join(x))
tv_df['Resolution']=tv_df['Resolution'].apply(lambda x:''.join(x))
```

```python
tv_df['Price']=tv_df['Price'].apply(lambda x:x.replace('₹','').replace(',',''))
```

#### OBSERVATIONS:
#### We, extracted supporting_apps and resolution of TV's.
#### And, we replaced rupee symbol and commas with nothing in Price column.
